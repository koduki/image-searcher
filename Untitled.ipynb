```ruby
require 'nokogiri'

html = Nokogiri::HTML.parse(open('sample01.html', "r:UTF-8"), nil, 'utf-8')
```

```ruby
puts html
html.xpath('//body').children[3].name

def f node
  xs = []
  node.element_children.each do |c|
    xs << f(c)
  end
  [[node.name, node.element_children.size], xs]
end

tree = f html
```
